```python
import sys
!{sys.executable} -m pip install pulp 
import pulp
from pulp import * 
prob = LpProblem('student_diet', LpMaximize)
```

```python
# Restaurant chosen = 1, not chosen = 0
x = LpVariable.dicts("visit", Restaurants, lowBound = 0, upBound = 1, cat = 'Integer')
```

```python
# Objective Function 
prob += lpSum([(calorie[r]/cost[r])*x[r] for r in Restaurants]) # maximize total restaurants average calorie intake per cost

# Constraints: 
prob += lpSum([calorie[r]*x[r] for r in Restaurants]) >= 1600 # meeting lowerbound of calorie intake requirement
prob += lpSum([calorie[r]*x[r] for r in Restaurants]) <= 3200 # not exceeding upperbound of calorie intake requirement
prob += lpSum([cost[r]*x[r] for r in Restaurants]) <= rmp_daily # cost not exceeding that of the UBC RMP
prob += lpSum([x[r] for r in Restaurants]) == 3 # 3 restuarants for 3 meals

print(prob)
```

```python
prob.solve()
print("Status:", LpStatus[prob.status])
```

```python
print("Restaurants:")
for a in prob.variables():
    if a.varValue == 1:
        print(a)

print("Restaurants' average calories:")
for a in prob.variables():
    if a.varValue == 1:
        print(a, ":", calories[a])

print("Restaurants' average cost:")
for a in prob.variables():
    if a.varValue == 1:
        print(a, ": $", round(cost[a], 2))
        
print("Restaurants' average calories per cost:")
for a in prob.variables():
    if a.varValue == 1:
        print(a, ":", calories[a]/cost[a])

daily_cost = 0;
for a in prob.variables():
    if a.varValue == 1:
        daily_cost += cost[a]
print("Average daily cost for all 3 meals = $", round(daily_cost, 2))

daily_calories = 0;
for a in prob.variables():
    if a.varValue == 1:
        daily_calories += calories[a]
print("Average daily calories for all 3 meals = $", daily_calories)

print("Money saved from UBC RMP by average daily cost =", round(rmp_daily - daily_cost, 2))
print("Daily calories per daily cost spent =", "$", round(daily_calories/daily_cost, 2))
```
